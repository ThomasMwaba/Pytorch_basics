```python
import numpy as np
import torch
```

#### Tensor Arithmetic

```python
a = torch.Tensor([1,2,3]) # creates a tensor a with floats
```

```python
b = torch.Tensor([4,5,6]) # creates b tensor a with floats
```

```python
print(a)
print(b)
```

###### Addition

```python
a + b # adds tensor a and tensor b
```

```python
torch.add(a,b) # adds tensor a and tensor b
```

###### subtraction

```python
a - b # removes tensor b from tensor a
```

```python
torch.sub(a,b) # removes tensor b from tensor a
```

###### Multiplication

```python
a * b # multiplies tensor a by tensor b
```

```python
torch.mul(a,b) # multiplies tensor a by tensor b
```

##### Reassignment

```python
a.add_(b) # reassigns to a = a + b 
```

```python
a.mul_(b) # reassigns to a = a * b 
```

```python
a.dot(b) # dot product of a and b
```
